```python
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
import cv2
import numpy as np
from PIL import Image

# Load processor dan model TroCR
processor = TrOCRProcessor.from_pretrained("microsoft/trocr-base-handwritten")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-base-handwritten")

# Fungsi preprocessing
def preprocess_image(cropped_image):
    # Resize untuk meningkatkan resolusi
    cropped_image = cv2.resize(cropped_image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

    # Grayscale
    gray_image = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2GRAY)

    # Denoising
    denoised_image = cv2.medianBlur(gray_image, 3)

    # CLAHE (Contrast Limited Adaptive Histogram Equalization) untuk kontras lebih baik
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    enhanced_image = clahe.apply(denoised_image)

    # Thresholding untuk membuat gambar menjadi biner
    _, binary_image = cv2.threshold(enhanced_image, 150, 255, cv2.THRESH_BINARY)

    # Sharpening untuk meningkatkan ketajaman teks
    kernel = np.array([[0, -1, 0],
                       [-1, 5, -1],
                       [0, -1, 0]])
    sharpened_image = cv2.filter2D(binary_image, -1, kernel)

    return sharpened_image

# Loop untuk tiap bounding box
for idx, box in enumerate(results[0].boxes):
    x1, y1, x2, y2 = map(int, box.xyxy[0])
    cropped_image = image[y1:y2, x1:x2]  # Potong gambar sesuai bounding box

    # Preprocess gambar agar lebih tajam dan jelas
    preprocessed_image = preprocess_image(cropped_image)

    # Konversi ke format PIL untuk TroCR
    pil_image = Image.fromarray(cv2.cvtColor(preprocessed_image, cv2.COLOR_BGR2RGB))

    # Preprocess dan ekstrak teks
    pixel_values = processor(images=pil_image, return_tensors="pt").pixel_values
    generated_ids = model.generate(pixel_values)
    text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

    # Output teks per bounding box
    print(f"Bounding Box {idx+1}: {text}")
```
